## Date Conversion

Convert the Order Date and Ship Date columns from string format to datetime format for time-series analysis.

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="%m/%d/%Y"
)

df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    format="%m/%d/%Y"
)

df.dtypes
```

## Feature Engineering

Create additional date-related features such as:

- Order Year
- Order Month
- Order Month Name
- Order Quarter

```python
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month
df["Order Month Name"] = df["Order Date"].dt.month_name()
df["Order Quarter"] = df["Order Date"].dt.quarter

df.head()
```

## Save Cleaned Dataset

Save the cleaned dataset into the processed folder for use in future notebooks.

```python
df.to_csv("../data/processed/clean_superstore.csv", index=False)

print("Cleaned dataset saved successfully!")
```
